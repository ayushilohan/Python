##### Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different

```python
## creating a tuple

empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))

tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
print(numbers)
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuples=(1,"hello world", 3.14, True)
print(mixed_tuples)
```

```python
## accessing Tuple elements
numbers
```

```python
print(numbers[2])
print(numbers[-1])

# indexing and slicing will be done similar to list
```

```python
numbers[0:4]
```

```python
numbers[::]
```

```python
numbers[::-1]
```

```python
## Tuple operations 

cont = numbers + mixed_tuples
print(cont)
```

```python
mixed_tuples * 3
```

```python
numbers* 3
```

```python
## immutable nature of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned

lst=[1,2,3,4,5]
print(lst)

lst[1]="ayushi"
print(lst)
```

```python
numbers[1]="ayushi"
```

```python
## Tuple methods

print(numbers)
print(numbers.count(1)) # count occurance of 1 in tuple
print(numbers.index(3)) # in which index 3 is present
```

```python
## packing and unpacking

## packing
packed_tuple=1, "hello", 3.14
print(packed_tuple)
```

```python
## unpacking
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## unpacking with *

numbers=(1,2,3,4,5,6)

first, *middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
## nested Tuple

## nested List
lst=[[1,2,3,4], [6,7,8,9,], [1,"hello",3.14,"c"]]
print(lst[0])
print(lst[0][2])
print(lst[0][0:3])
```

```python
lst=[[1,2,3,4], [6,7,8,9,], (1,"hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple = ((1,2,3), ("a", "b", "c"), (True, False))

## access the elements inside a Tuple

print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested Tuple

for sub_tuple in nested_tuple:
    for items in sub_tuple:
        print(items, end="")
    print()
```
