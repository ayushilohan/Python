#### Dictionaries are unordered collections of items. They store data in key value pairs. Keys must be unique and immutable/cannot be changed (eg. strings, numbers or tuples), while values can be of any type

```python
## create dictionary
# [] - list
# () - tuples

empty_dic={}
print(type(empty_dic))
```

```python
# create dictionary using dict class

empty_dic=dict()
print(type(empty_dic))
```

```python
student={
    "name": "ayushi",
    "age": 25,
    "grade": 24
}
print(student)
print(type(student))
```

```python
# error - single key is always used

student={
    "name": "ayushi",
    "age": 25,
    "name": 24 # replace the value
}

print(student)
```

```python
## accessing dictionary elements

student={
    "name": "ayushi",
    "age": 25,
    "grade": 'A'
}
print(student['name'])
print(student['age'])
print(student['grade'])
```

```python
## acessing using get() method

print(student.get('name'))
print(student.get('grade'))

print(student.get('last_name'))
print(student.get('last_name', "not available"))
```

```python
## modifying dictionary elements
## dictionaries are mutuable, you can add, update or delete elements

print(student)
```

```python
# update value for the key

student["age"]=33
print(student)

# add new key and value

student["address"]="india"
print(student)
```

```python
# delete the key

del student['grade'] # delete key and value pair
print(student)
```

```python
# dictionary methods

keys=student.keys()
values=student.values()

print(keys)
print(values)

items=student.items() # get all key values pairs
print(items)
```

```python
## shallow copy

student_copy=student # refer to student in memory
print(student)
print(student_copy)
```

```python
# problem

student["name"]="sunita"
print(student)
print(student_copy)
```

```python
# solution

student_copy1=student.copy() # allocate a different memory with same values
print(student)
print(student_copy1)
```

```python
student["name"]="ayushi"
print(student)
print(student_copy1)
```

```python
# iteration over dictionary

# can use loops to iterate over dictionaries, keys, values or items

# iterate over keys

for keys in student.keys():
    print(keys)

# iterate over values

for values in student.values():
    print(values)
```

```python
# iterate over key value pairs

for key,value in student.items():
    print(f"key : {key} and value : {value}")
```

```python
##  Nested dictionaries

students={
    "student1": {"name": "ayushi", "age": 25},
    "student2": {"name": "sunita", "age": 50}
}
print(students)
```

```python
## access nested dictionaries elements

print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## iterating over nested dictionaries

for student_id,student_info in students.items():
    print(f"{student_id} : {student_info}")
    for key,value in student_info.items():
        print(f"{key}: {value}")
```

```python
## dictionary comprehension

squares={ x:x**2 for x in range(5) }
print(squares)
```

```python
## conditional dictionary comprehension

evens= { x:x**2 for x in range(10) if x%2==0 }
print(evens)
```

```python
## practical examples

# use a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,5,3,4,4,4,4]
result={}

for key in numbers:
    if key in result:
        result[key] +=1
    else:
        result[key] = 1

print(result)
```

```python
## merge 2 dictionaries into one

dict1={"a": 1, "b": 2}
dict2={"c": 3, "d": 4}

merged_dict={**dict1, **dict2}
print(merged_dict)
```
