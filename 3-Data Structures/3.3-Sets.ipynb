```python
# create a set
my_set={1,2,3,4,5}
print(type(my_set))
print(my_set)
```

```python
my_empty_set=set()
my_empty_set1=set([1,2,3,4,5])

print(type(my_empty_set))
print(my_empty_set)

print(type(my_empty_set1))
print(my_empty_set1)
```

```python
my_set=set([1,2,3,4,6,5,6,4])
print(my_set) # doesnt print duplicate values
```

```python
# basic set operation

# add elements

my_set.add(7)
print(my_set)

my_set.add(7)
print(my_set)
```

```python
# remove elements

my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
print(my_set)
```

```python
my_set.discard(11) # remove element from the set if present and does not throw error if element not present in set
print(my_set)
```

```python
removed_element=my_set.pop() # first in first out
print(removed_element)
print(my_set)
```

```python
# clear all the elements

my_set.clear()
print(my_set)
```

```python
## set membership test

# whether a element is member of the set

my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
# mathematical operation

set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

# union - combine everything
union_set=set1.union(set2)
print(union_set)

#intersection - common elements

intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2) # common elements got updated in set1
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

# difference
print(set1.difference(set2)) # from set 1 remove common elements from set2
print(set1)
```

```python
print(set2.difference(set1))
print(set2)
```

```python
# symmetric difference

print(set1.symmetric_difference(set2)) # common element removed from both sets
print(set1)
print(set2)
```

```python
# sets method
set1={1,2,3}
set2={3,4,5}

## is subset

print(set1.issubset(set2)) # is all elements from set 1 available in set 2
print(set1.issuperset(set2)) # whether set 1 has all elements of set 2
```

```python
## counting unique words in text

text="in this tutorial we are discussing about sets"
words=text.split()
print(words)

## convert list of words to set to get unique words

unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
