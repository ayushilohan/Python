### Example 1: Manage a To Do list
#### Create a to do list to keep track of tasks

```python
to_do_list=["buy grocery", "clean the house", "pay bills"]

# add the task
to_do_list.append("schedule a meeting")
to_do_list.append("go for run")

# remove the task
to_do_list.remove("clean the house")

# check if a task is in last
if "pay bills" in to_do_list:
    print("dont forget to pay bills")

print("remaing task in the list")

for item in to_do_list:
    print(f"my task : {item}")
```

### Example 2: Organizing Student Grades
#### Create a list to store an calculate average grades for students

```python
grades=[85, 92, 78, 90, 88]

grades.append(95)

average_grade = sum(grades) / len(grades)
print(f"average grade : {average_grade:.2f}")

highest_grade = max(grades)
lowest_grade = min(grades)

print(f"highest grade : {highest_grade}")
print(f"lowest grade : {lowest_grade}")
```
