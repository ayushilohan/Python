```python
num = int(input("enter number : "))

if num>=0:
    print("number is positive")
    if num%2==0:
        print("number is even")
    else:
        print("number is odd")
else:
    print("number is negative")
```

```python
year = int(input("Enter year : "))

if year>=0:
    print("Number is positive")
    if year%4==0:
        print("It is leap year")
    else:
        print("It is not leap year")
else:
    print("Number not valid")
```
